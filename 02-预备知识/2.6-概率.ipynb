```python
%matplotlib inline
import torch
from torch.distributions import multinomial
from d2l import torch as d2l
```

```python
fair_probs = torch.ones([6]) / 6
multinomial.Multinomial(1, fair_probs).sample()
```

```python
multinomial.Multinomial(10, fair_probs).sample()
```

```python
counts = multinomial.Multinomial(1000, fair_probs).sample()
counts / len(counts)
```

```python
counts = multinomial.Multinomial(10, fair_probs).sample((500,))
cum_counts = counts.cumsum(dim=0)
estimates = cum_counts / cum_counts.sum(dim=1, keepdim=True)

d2l.set_figsize((6, 4.5))
for i in range(6):
    d2l.plt.plot(estimates[:, i].numpy(), label=('P(die=)' + str(i+1) + ")"))

d2l.plt.axhline(y=0.167, color='black', linestyle='dashed')
d2l.plt.gca().set_xlabel("Group of experiments")
d2l.plt.gca().set_ylabel("Estimated probability")
d2l.plt.legend()
```

```python
diff = []
for m in range(100, 1000):
        counts = multinomial.Multinomial(100, fair_probs).sample((m,))
        cum_counts = counts.cumsum(dim=0)
        estimates = (max(cum_counts[-1])  - min(cum_counts[-1]))/ cum_counts[-1].sum()
        diff.append(estimates)

x = range(100, 1000)
d2l.plt.plot(x, diff)
```

```python
diff = []
for n in range(1, 20):
        counts = multinomial.Multinomial(n, fair_probs).sample((500,))
        cum_counts = counts.cumsum(dim=0)
        estimates = (max(cum_counts[-1])  - min(cum_counts[-1]))/ cum_counts[-1].sum()
        diff.append(estimates)

x = range(1, 20)
d2l.plt.plot(x, diff)
```

```python
# 创建数据  
x = []
y = []
z = []
for m in range(100, 800, 10):
        for n in range(10, 100, 5):
            counts = multinomial.Multinomial(n, fair_probs).sample((m,))
            cum_counts = counts.cumsum(dim=0)
            estimates = (max(cum_counts[-1])  - min(cum_counts[-1]))/ cum_counts[-1].sum()
            x.append(m)
            y.append(n)
            z.append(estimates)
```

```python
# 创建3D图形对象  
fig = d2l.plt.figure()  
ax = fig.add_subplot(111, projection='3d')  
  
# 绘制散点图  
ax.scatter(x, y, z, c='r', marker='.')  
  
# 设置坐标轴标签和标题  
ax.set_xlabel('m')  
ax.set_ylabel('n')  
ax.set_zlabel('diff')  
ax.set_title('Max diff')  
```
